```python
import pandas as pd
```

```python
data = pd.read_csv('50_Startups.csv')
```

```python
data.head()
```

```python
X = data.drop(['Profit'], axis = 1)
```

```python
y = data['Profit']
```

```python
y.head()
```

**First dummy variable is removed to avoid dummy trap as in** 
**this trap it create a trap in which minimum one become dependent**
**so removing one might solve the problem in various situations**

```python
dummy = pd.get_dummies(data['State'], prefix = "State_", drop_first= True)
```

```python
X.drop(['State'], axis = 1, inplace = True)
```

```python
dummy.head()
```

```python
X = pd.concat([X.copy(), dummy.copy()], axis=1)
```

```python
X.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.predict([[16578, 136749, 284682, 0, 1]])
```
