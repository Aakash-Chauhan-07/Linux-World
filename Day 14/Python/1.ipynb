```python
class VisitorForm:
    name = None
    phone = None
    email = None
    query = None

    def __init__(self, name, phone, email, query):
        self.name = name
        self.phone = phone
        self.email = email
        self.query = query

    def lw(self): # Self is standard , otherwise its just a variable that take address of object
        print("I am LW")
        print(self.name)
        print(self.phone)
        print(self.email)
        print(self.query)
```

```python
tom = VisitorForm(name = 'tom',
                  phone = 123,
                  email = "tom@gmail.com",
                  query = "Time pass")
```

```python
# By function in class take one paramter
```

```python
tom.lw()
```
